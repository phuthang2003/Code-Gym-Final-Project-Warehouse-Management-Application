```python
import turtle

screen = turtle.Screen()
screen.setup(width=800, height=600)

t = turtle.Turtle()
t.speed(1)

def draw_house():
    # Draw the house
    t.penup()
    t.goto(-50, -50)
    t.pendown()
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)

    # Vẽ ngói
    t.left(45)
    t.forward(70)
    t.right(90)
    t.forward(70)
    t.right(135)
    t.forward(100)

def draw_tree(x, y, size):
    # Draw the thân cây
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.color("gray")
    t.begin_fill()
    t.forward(size/4)
    t.left(90)
    t.forward(size)
    t.left(90)
    t.forward(size/4)
    t.left(90)
    t.forward(size)
    t.left(90)
    t.end_fill()

    # Draw the lá cây
    t.color("green")
    for i in range(3):
        t.begin_fill()
        t.forward(size/2)
        t.left(120)
        t.forward(size/2)
        t.left(120)
        t.forward(size/2)
        t.left(120)
        t.end_fill()
        t.forward(size/4)
        t.right(90)
        t.forward(size/4)
        t.left(90)
    
# Vẽ nhà
draw_house()

# Draw cây
draw_tree(-150, -100, 100)
draw_tree(100, -100, 150)

# Ẩn turtle
t.hideturtle()
turtle.done()
```
