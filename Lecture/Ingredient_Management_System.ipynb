```python
class Ingredient:

    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
    def __str__(self):
        return f"{self.name}, {self.quantity}"
```

```python
from tkinter import *
from tkinter import messagebox
# Tạo ra cửa số ứng dụng
window = Tk()

# Setup thôn tin cho window
window.title("Ingredients Management System")
window.geometry("350x150")

# Tạo list để lưu trữ toàn bộ các objective

list_ingredients = []

# Mô tả các function cho ứng dụng
# add/search/update/delete/reset
def add_ingredient():
    # Lấy giá trị từ các entry
    name_value = name.get()
    quantity_value = int(quantity.get())

    new_ingredient = Ingredient(name_value, quantity_value)

    # Thêm vào list
    list_ingredients.append(new_ingredient)

    # List comprehension
    print([item.name for item in list_ingredients])
    print([item.quantity for item in list_ingredients])

    # Hiện thị thông báo
    messagebox.showinfo("Add Ingredient", f"{name_value} is added successfully.")

    # xóa giá trị ở entry
    name.delete(0, END)
    quantity.delete(0, END)

def search_ingredient():
    search_value = search_name.get()
    is_exist = False

    for ingredients in list_ingredients:
        if ingredients.name == search_value:
            is_exist = True
            messagebox.showinfo("Ingredient", f"Name:{ingredients.name}, quantity: {ingredients.quantity}.")
            break
    
    if is_exist == False:
        messagebox.showinfo("Search Ingredient", f"{search_value} is not found.")

def delete_ingredient():
    search_value = search_name.get()
    is_exist = False

    for ingredients in list_ingredients:
        if ingredients.name == search_value:
            is_exist = True
            # Xóa ingredient khỏi list
            list_ingredients.remove(ingredients)
            messagebox.showinfo("Delete Ingredient", f"{ingredients.name} is deleted successfully")
            break
    
    if is_exist == False:
        messagebox.showinfo("Delete Ingredient", f"{search_value} is not found.")

def update_ingredient():
    update_name = name.get()
    update_quantity = int(quantity.get())
    is_exist = False

    for ingredients in list_ingredients:
        if ingredients.name == update_name:
            is_exist = True
            # Thay đổi quantity
            list_ingredients.remove(ingredients)
            update_ingredient = Ingredient(update_name, update_quantity)
            list_ingredients.append(update_ingredient)
            messagebox.showinfo("Update Ingredient", f"{update_name}, quantity: {update_quantity}.")
            print([item.name for item in list_ingredients])
            print([item.quantity for item in list_ingredients])
            break
    
    if is_exist == False:
        messagebox.showinfo("Update Ingredient", f"{update_name} is not found.")
    pass

def reset():
    name.delete(0, END)
    quantity.delete(0, END)
    search_name.delete(0, END)


# Setup các widget
# Tạo label
name_lbl = Label(window, text = "Name: ")
quantity_lbl = Label(window, text="Quantity: ")

# Tạo entry
name = Entry(window)
quantity = Entry(window)
search_name = Entry(window)

# Tạo button
add_btn = Button(window, text="Add", command = add_ingredient)
search_btn = Button(window, text="Search", command = search_ingredient)
update_btn = Button(window, text="Update", command=update_ingredient)
delete_btn = Button(window, text="Delete", command= delete_ingredient)
reset_btn = Button(window, text="Reset", command= reset)

# Dán các widget lên window
name_lbl.grid(row = 0, column = 0)
quantity_lbl.grid(row = 1, column = 0)
add_btn.grid(row = 2, column = 0)
name.grid(row = 0, column = 1)
quantity.grid(row = 1, column = 1)
search_name.grid(row = 3, column = 0)
search_btn.grid(row = 3, column = 1)
update_btn.grid(row = 3, column = 2)
delete_btn.grid(row = 2, column = 1)
reset_btn.grid(row = 2, column = 2)

# Chạy ứng dụng
window.mainloop()
```
